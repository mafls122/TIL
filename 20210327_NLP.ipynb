```python
import nltk
from konlpy.tag import Kkma
from konlpy.tag import Okt
from collections import Counter
```

```python
# 명사 분석 (nouns)

f = open('문재인.txt','r',encoding='utf-8')
news = f.read()

# okt 객체 생성
okt = Okt()

# 명사 추출
noun = okt.nouns(news)
for i,v in enumerate(noun):
    # 1글자 삭제
    if len(v) < 2 :
        noun.pop(i)
            
# 명사 빈도 카운트 및 저장(dict)        
count = Counter(noun)
# print(count)

# 명사 빈도 top100
noun_list = count.most_common(100)
print(noun_list)
# for v in noun_list:
#     print(v)
```

```python
# 형태소 분석 (morphs)

f = open('문재인.txt','r',encoding='utf-8')
news = f.read()

# okt 객체 생성
okt = Okt()

# 명사 추출
morph = okt.morphs(news)
            
# 명사 빈도 카운트 및 저장(dict)        
count = Counter(morph)
# print(count)

# 명사 빈도 top100
morph_list = count.most_common(100)
print(morph_list)
# for v in noun_list:
#     print(v)
```

```python
from wordcloud import WordCloud

wc = WordCloud(font_path='C:\\Users\\user\\han.ttf', 
    background_color="white", width=1000, height=1000, max_words=100, max_font_size=300)

wc.generate_from_frequencies(dict(morph_list))
# wc.to_file('wordcloud_news.png')  # 파일 저장

pyplot.figure(figsize=(12, 9))
pyplot.imshow(wc)
pyplot.axis("off")
pyplot.show()
```
