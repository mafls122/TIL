- 2. 좋은 직선을 찾는 기준, 평균 제곱근 오차(RMSE)

평균기온($X$ | 아이스크림 쇼핑 클릭량($Y$
---|---
5  | 16
10 | 23
15 | 31
20 | 50

직선의 방정식 $f_1(X)= 2X+5$ 라는 생각으로 오차를 구해보자

오차값 = 실제값($Y$) - 예측값($f_1(X)$)

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([5,10,15,20])  # 평균기온 X
Y = np.array([16,23,31,50])  # 실제 쇼핑 클릭량 Y

f1_x = 2*X + 5
f1_error = Y - f1_x
print(f1_error)
print(np.sum(f1_error))
```

 입력변수 | 예측값 | 실제값 | 오차
 --- | ---  | ---  | --- |
 5   |   15 | 16   |  1  |
 10  |   25 | 23   | -2  |
 15  |   35 | 31   | -4  |
 20  |   45 | 50   |  5  |

```python
# 그래프, 예측치의 직선 방정식
plt.plot(X, f1_x)

#오차값 그리기
for i in range(len(X)):
    x = np.array([X[i], X[i]])  # 현재 온도
    y = np.array([f1_x[i], Y[i]])  # 실제 값과 예측값
    plt.plot(x,y, marker='o')
    
plt.grid()
plt.show()
```

```python
# 직선의 방정식 f2(X) = 30 라고 생각할 경우 오차를 구하자
# 오차의 합은 0, 수평 직선은 데이터가 제대로 표현되지 못하고 있다.

f2_x = np.full((X.shape),30)
f2_error = Y-f2_x
print(f2_error)
print(np.sum(f2_error))
```

```python
# 새로운 예측값을 지정 y=30, 오차의 합은 0
# 그래프, 예측치의 직선 방정식

plt.plot(X,f2_x)

#오차값 그리기
for i in range(len(X)):
    x = np.array([X[i], X[i]])  # 현재 온도
    y = np.array([f2_x[i], Y[i]])  # 실제 값과 예측값
    plt.plot(x,y, marker='o')
    
plt.grid()
plt.show()
```

```python
# f3(X) = -2X+ 55 라고 생각하고 오차를 구해보자

f3_x = -2 * X + 55
f3_error = Y - f3_x

print(f3_error)
print(np.sum(f3_error))
```

```python
# 새로운 예측값을 지정 -2x + 55, 오차의 합은 0
# 그래프, 예측치의 직선 방정식

plt.plot(X,f3_x)

#오차값 그리기
for i in range(len(X)):
    x = np.array([X[i], X[i]])  # 현재 온도
    y = np.array([f3_x[i], Y[i]])  # 실제 값과 예측값
    plt.plot(x,y, marker='o')
    
plt.grid()
plt.show()
```

```python
# 3개의 직선을 모두 그려 보자

plt.scatter(X,Y)
plt.plot(X,f1_x)
plt.plot(X,f2_x)
plt.plot(X,f3_x)
plt.grid()
plt.show()
```

정답과 예측치와의 오차의 단순 합은 모두 0으로 만들어 사용하기 때문에 어떤 공식을 적용해도 결과 데이터 혹은 데이터 표시가 어떤 것이 더 좋은지 판단할 수 없다.

음수의 값으로 예측값보다 실제 값이 작은 값이 발생할 경우가 생겨난다.

이 원인을 가지고 오차의 결과를 양수로 만들어 주는 작업을 하고 새롭게 제곱근들의 평균을 구한 후 이 값의 제곱근의 오차를 구한다.

RMSE = Root Mean Square Error
$$
RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}{(Y_i - f(X_i))^2}}
$$

```python
def RMSE(a,b,X,y):
    RMSE = np.sqrt( ( (y-(a*X+b))**2 ).mean() )
    return RMSE

# 처음 지정한 공식 f(x) = 2x+5  
# 첫번째 방식이 에러가 가장 작은값을 리턴하기 때문에 데이터의 특성을 가장 잘 살렸다고 판단
print(RMSE(2,5,X,Y))

# 처음 지정한 공식 f(x) = 30
print(RMSE(0,30,X,Y))

# 처음 지정한 공식 f(x) = -2x+55
print(RMSE(-2,55,X,Y))
```

평균제곱근오차와 같이 좋은 직선을 찾기 위해 크기를 최소화해야 하는 값에 대한 함수를 선형회귀 모델의 비용함수(Cost Function)라고 한다.

$$
f(X) = \beta_1X + \beta_0
$$
$$
Cost(\beta_0, \beta_1) = \sqrt{\frac{1}{n}\sum_{i=1}^{n}{(Y_i - f(X_i))^2}}
$$

```python
# 온도와 아이스크림 클릭량을 이용해서 데이터 관계를 살펴보자
# 1. Data 행렬로 구현해서 출력
import csv

f = open('res_ice.csv', encoding='euc-kr')
data = csv.reader(f)
header = next(data) # 헤더 처리

temp=[]  # 평균 기온 데이터 저장소
ice=[]  # 아이스크림 판매 데이터 저장

for row in data:
    temp.append(float(row[1]))
    ice.append(int(row[4]))

print(temp)
print(ice)
```

```python
# 2. 평균 기온과 아이스크림의 최대값과 최소값을 출력해보자

print(max(temp),min(temp))
print(max(ice),min(ice))

# 3. 평균 기온값을 기준으로 도수 분포표를 생성 ( 온도, 도수구간 ) 히스토그램 그리기
import numpy as np

bins = np.arange(min(temp),max(temp)+5,5)  # 5개만 출력, 구간을 늘려주는 것
bins

hist, bins = np.histogram(temp, bins) # 히스토그램 계산 -> 총 데이터, bins 계급값
print(hist)
print(bins)

# 4. 평균 기온 기준으로 아이스크림 구분해서 저장
ice_buy = np.zeros(7)
ice_buy

for i in range(0,len(temp)): # 평균 기온 데이터 전체 탐색
    if bins[0] <= temp[i] and temp[i]<bins[1] : # 첫번째 인덱스에 해당한다면
        ice_buy[0] = ice_buy[0] + ice[i]  # 첫번째 배열값에 해당하는 날짜의 아이스크림 클릭량을 저장
    elif bins[1] <= temp[i] and temp[i]<bins[2]:
        ice_buy[1] = ice_buy[1] + ice[i]
    elif bins[2] <= temp[i] and temp[i]<bins[3]:
        ice_buy[2] = ice_buy[2] + ice[i]
    elif bins[3] <= temp[i] and temp[i]<bins[4]:
        ice_buy[3] = ice_buy[3] + ice[i]
    elif bins[4] <= temp[i] and temp[i]<bins[5]:
        ice_buy[4] = ice_buy[4] + ice[i]
    elif bins[5] <= temp[i] and temp[i]<bins[6]:
        ice_buy[5] = ice_buy[5] + ice[i]
    else :
        ice_buy[6] = ice_buy[6]+ice[i]
        
print(ice_buy)

# 이중 for로 구현
ice_buy02 = np.zeros(7)

for i in range(0,len(temp)):
    for j in range(0, len(bins)):
        if bins[j] <= temp[i] and temp[i] < bins[j+1] :
            ice_buy02[j] = ice_buy02[j] + ice[i]

print(ice_buy02)
```

```python
# 5. 계급별 아이스크림 쇼핑 클릭량의 평균을 구하자
# 평균 기온값 계급별 아이스크림 쇼핑 클릭량을 구하자

ice_buy_a = np.zeros(7)

for i in range(0,len(ice_buy)):
    ice_buy_a[i] = ice_buy[i] / hist[i] # 계급별 아이스크림 쇼핑 클릭량의 평균
    
# 아이스크림 클릭량의 평균을 소수 이하 2자리까지 출력하자
for i in range(0,len(ice_buy)):
    print("%.2f" %ice_buy_a[i])
    
# 6. 평균 기온에 따른 아이스크림 클릭량의 평균을 막대 그래프로 그리자
import matplotlib.pyplot as plt

print(bins)
plt.xlabel('avg temp')
plt.ylabel('num if ice shopping')
plt.bar(bins[0:7], ice_buy_a, width=2, align='edge')
plt.xticks(bins[0:7])
plt.show()
```

```python
# 7. 데이터 학습, 평가 데이터 분리하기
# 데이터 로드
import pandas as pd

df = pd.read_csv('res_ice.csv', encoding='euc-kr')
df

# 7-1. 학습 데이터의 입출력 변수 나누기
data = np.array(df)
data

X = data[:,1]  # 평균기온
y = data[:,4] # ==data[:,-1]

# 7-2. 비용을 계산하고 업데이트
mean_x = np.mean(X)
mean_y = np.mean(y)

# X 변수가 가진 개수
n = len(X)

# 최소 제곱법 beta0, beta1
temp01 = 0
temp02 = 0

for i in range(n):
    temp01 += (X[i]-mean_x)*(y[i]-mean_y) # 최소제곱법 공식
    temp02 += (X[i]-mean_x)**2
    
beta1 = temp01 / temp02
beta0 = mean_y - (beta1 * mean_x)

print("기울기 : ",beta1)
print("절편 : ",beta0)

# 7-3. 시각화
y_pred = beta0 + beta1 * X
plt.title("avg_temp, Clicks")
plt.xlabel('avg temp')
plt.ylabel('clicks')
plt.scatter(X,y)
plt.plot(X,y_pred, color='black')
plt.axis([-4,30,20,100])
plt.grid()
plt.show()
```

```python
# 데이터 학습, 평가 데이터로 분리
import pandas as pd

df = pd.read_csv('res_ice.csv',encoding='euc-kr')
df

# 학습 데이터를 입력 변수, 출력 변수 나누기
data = np.array(df)
X = data[:,1]
y = data[:,-1]

# 비용 계산하고 업데이트
mean_x = np.mean(X)
mean_y = np.mean(y)

# X 변수가 가진 개수
n = len(X)

# 최소 제곱법 beta0, beta1
temp01 = 0
temp02 = 0

for i in range(n):
    temp01 += (X[i]-mean_x)*(y[i]-mean_y) # 최소제곱법 공식
    temp02 += (X[i]-mean_x)**2
    
beta1 = temp01 / temp02
beta0 = mean_y - (beta1 * mean_x)

print(beta1)
print(beta0)

# 오차율 -> 손실금액 평가하기
def RMSE(beta0, beta1, x, y):
    RMSE = np.sqrt( ((y-(beta0+beta1*1*x))**2).mean() )
    return RMSE

res = RMSE(beta0,beta1, X,y)
print("손실값 결과는? {}".format(res))

# 문제 해결
def My_Regression(beta0, beta1,X):
    y_pred = beta0+beta1*X
    return y_pred

my_res = float(input("오늘의 기온을 입력 : "))

predicted_value = My_Regression(beta0,beta1,my_res)
print("오늘 아이스크림 클릭수는 100점을 기준으로{}만큼 예상".format(predicted_value))

               
```
