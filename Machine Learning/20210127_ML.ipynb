8. ML sklearn.metrics의 메소드를 살펴 보자

8-1 confusion matrix 형태의 데이터를 관리한다.
   -> 클래스 분류 결과를 "실제 클래스"와 "예측 클래스"를 축으로 가진 형태
     실제클래스, 예측 클래스( 두 값의 분류 ) -> 4개의 종류로 나뉜다.
                          실제 클래스                  예측 클래스
    TP(true positive)          T                           T      (정답)
    TN(true negative)          F                           F      (정답)
    FP(false positive)         F                           T      (오답)
    FN(false negative)         T                           F      (오답)
    
    
   case 1 =================================> Predicted
                                       negative   positive
    Actual                  negative     TN         FP
                            positive     FN         TP
                            
   case 2 =================================> 0 = Negative  1 = Positive
                                          0          1
    Actual                    0          TN         FP
                              1          FN         TP
                              
   case 3 =================================> A = Negative  B = Positive
                                          A          B
    Actual                    A          TN         FP
                              B          FN         TP
                              
   case 4 =================================> B = Negative  A = Positive
                                          A          B
    Actual                    A          TP         FN
                              B          FP         TN
                           -> TP,FN,FP,TN = cm.flatten()
                           
                  
   case 5 =================================> 
               0 = positive  1 = negative  labels = [0,1,2]  
               
                                         0          1          2
    Actual                    0          TP         FN         FN
                              1          FP         TN         TN
                              2          FP         TN         TN
                              
    
8-2 confusion_matrix()를 생성해서 사용한다.

sklearn.metrics.confusion_matrix(y_true, y_pred, *, labels=None, sample_weight=None, normalize=None)

8-3 accuracy_score() 정답률

  ( TP + TN + FP + FN ) / (TP + TN)
 
8-4 적합율 ( PPV : positive predictive value) _ presicion_score()

   precision = (TP+FP) / TP
   FP가 커지면 적합율은 작아진다

8-5 재현율 recall_score()

   recall =  TP / ( TP + FN )
   FN이 커지면 재현율은 작아진다
  
8-6 F1 measure 값 f1_score()
 
   case 1 : precision + recall / 2 * precision * recall
   case 2 : 2 *  TP + FP + FN / 2 * TP
   
8-7 매크로 평균, 마이크로 평균, 가중 평균 precision_score()의 옵션을 사용한다.


9. train_test_split() : ndarray 객체 및 목록 등을 둘로 나누어 리턴하는 기능 함수
                        머신러닝 데이터를 훈련(학습)과 테스트로 분할한다.
                        
                        데이터 검증방법 3가지
                        1. 홀드 아웃 검증 : 데이터를 훈련 데이터와 테스트 데이터를 분할 한 다음 테스트 데이터를 사용하지 않는 방법
                        
                        2. 교차검증(n겹) : 데이터를 n개의 그룹으로 나누어서 원하는 그룹을 하나 선택한 후 
                                           나머지 데이터를 훈련데이터로 사용한다. 원하는 그룹을 하나 선택한 것은 테스트 평가, 
                                           테스트 데이터에서 정밀도의 평균, 표준편차를 구해서 평가 실행 (5~10내외로 그룹을 나눈다)
                        
                        3. LOOCV(Leave-One-Out-Cross-Validation) : 교차 검증시에 그룹개수와 데이터수가 같을 경우 사용하는 검증 방법

< 기계학습의 흐름 >

데이터 수집 -> 전처리 -> 데이터 분할 -> 학습하기 -> 테스트 평가하기

1. 데이터 분할 : sklearn.model_selection.train_test_split : 목적 데이터를 분할

2. 학습하기 (다양한 패턴의 알고리즘) : sklearn.ensemb.RandomFroestClassfier 등
                                        sklearn.ensemble. *
 2-1. ensemble(집단학습) : 미학습 데이터에 대한 예측 능력을 평가하는 것
                           단일 학습, 융합
                           복잡한 학습 -> 독립적, 간단한 학습 -> 순차적 수행
                           RandomForest, Bootstrap 두가지 관점으로 나뉜다.
 
 2-2. RandomForest 는 무슨 알고리즘인가?
    1단계) 부트스트랩 표본 추출을 복원 추출 : 학습 데이터에서 n개의 샘플을 무작위로 선택 
    2단계) 부트스트랩의 표본에서 결정 나무를 생성 : 임의의 d개의 특징된 양을 추출 
            (목적함수에 따라 최적의 분할이 되는 특징량을 사용하여 노드를 분할하는 작업 : 카테고리 분할)
    3단계) 예측된 결과를 리턴 -> 레이블로 할당
    
    ex) 100명과 찍은 친구와의 사진이 있다. 
       그 중 홍길동, 정길동, 박길동을 임의로 추출해서 베프인지 아닌지를 판별
       홍길동은 2단계의 분류기를 거쳤더니 베프 레이블, 
       정길동, 박길동은 2단계 분류기를 거쳤더니 베프가 아니다.
       
       테스트 ->
       100명 이외의 친구 엔코아와 내가 사진을 찍어서 테스트를 했더니 베프가 아님

3. 테스트 평가 : sklearn.metrics.accuracy_score 등
                 sklearn.metrics. *
                 
* 입출력함수가 정해져 있기 때문에 데이터 전체 구현하기전에 체크
  model.fit(X,y), model.fit(X), model.predict(X) : 모델에서 테스트까지 해줌

```python
# 데이터를 가지고 학습 및 평가, 한꺼번에 한거!

# 1. 데이터를 로드
# 2. 학습과 테스트로 나눈다
# 3. 모델 지정 = sklearn에서 제공하는 알고리즘 종류
# 4. 학습
# 5. 테스트 및 평가

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier # 지표를 지정해서 n개의 결정나무 앙상볼 학습 : 정밀도를 구현할 때 사용
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()
data, target = iris.data, iris.target

data_train, data_test, target_train, target_test = train_test_split(data,target, 
                                                                    train_size=0.8, random_state=1)
# 앙상블 학습 : 집단적으로 분류할때 사용
model = RandomForestClassifier()

model.fit(data_train, target_train) # 학습 데이터로 학습을 시킨다

target_pred = model.predict(data_test)
print(accuracy_score(target_test, target_pred)) #정답률 계산 accuracy_score
```

```python
 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn import datasets

# 1. 데이터 로드
iris = datasets.load_iris()
data, target = iris.data, iris.target

# 2. 데이터 분할(학습 데이터, 테스트 데이터)
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=0)

# 3. 모델 지정
model = RandomForestClassifier()

# 4. 모델 학습
model.fit(X_train, y_train)

# 5. 테스트 및 평가
y_pred = model.predict(X_test)
print('정답률 : ', accuracy_score(y_test, y_pred))
```
